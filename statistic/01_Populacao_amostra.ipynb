# Amostragem





## O que é?
É um método que permite obter informações sobre a população com base nas estatísticas de um recorte ou subconjunto da população, sem a necessidade de "olhar" para todos os indvíduos, mas para uma **amostra**. 

Sendo assim, para investigar as estatísticas de uma população, a fim de diminuir os custos com tempo, recurso e processamento que uma tarefa do tipo censo gera, usa-se a amostragem. 

Vale lembrar que a amostra precisa ser aleatória e representativa, de modo a diminuir eventuais viéses.

**Amostragem Probabilística**

Neste método, cada elemento da população tem a mesma chance de ser selecionado, o que proporciona um subconjunto da população que seja representativo. 

**Amostra não probabilística**

Neste método, os elementos não têm a mesma chance de serem selecionados, o que resulta numa deficiência de representatividade e generalização para a população.

Temos que a nossa população referente ao dataset é de 32561 indivíduos. 

Para uma amostragem sistemática, precisamos de uma amostra de 100 indivíduos. Para tal, primeiramente devemos encontrar o passo de amostragem. Isto é, o número que representa o intervalo de seleção aleatória de um indivíduo e os seus subsequentes. 

---



```python
tam_populacao = len(dataset)
tam_populacao
```

```python
# Calcular o passo de amostragem
passo = tam_populacao // 100
print('Passo de amostragem: ', passo)
```

Temos então que o passo de amostragem é o valor inteiro 325. Contudo, devemos também encontrar o primeiro indivíduo da amostragem. 

O índice do indivíduo, ou o valor que representa a posição do indivíduo na população, deve ser entre 0 e 325, o qual também será selecionado aleatoriamente. 

```python
# Seleção do primeiro indivíduo
random.seed(2) # Semente de randomização para manter o mesmo valor aleatório a cada rodada
primeiro_individuo = random.randint(0, passo) # Sorteia um valor aleatório entre 0 e 325
primeiro_individuo
```

```python
# Segundo indivíduo
segundo_individuo = primeiro_individuo + passo # 28 + 325
segundo_individuo
```

Podemos fazer isso de forma automática com a função `np.arange()` da biblioteca Numpy. Para rodar a função, pricisamos do número ou variável que representa o 1º elemento, o tamanho da população e o passo de amostragem. 

A função irá retornar um array de números que representam os índices dos indivíduos amostrados sistematicamente da população Census.

```python
np.arange(primeiro_individuo, tam_populacao, passo)
```

### Função amostragem sistemática

```python
def amostra_sistematica(dataset, tam_amostra):
  """
  Função que recebe um dataset e um valor do tamanho da amostra desejada, calcula
  a amostra pelo método sistemático e retorna o dataframe amostrado. 
  """
  tam_populacao = len(dataset)
  passo = tam_populacao // tam_amostra
  random.seed(2) 
  primeiro_individuo = random.randint(0, passo)
  indices = np.arange(primeiro_individuo, tam_populacao, step = passo)
  amostra = dataset.iloc[indices]

  return amostra
```

```python
df_amostra_sistematica = amostra_sistematica(dataset, 100)
df_amostra_sistematica.head()
```

### Função amostragem por agrupamento

Faremos agora uma função para realizar a amostragem por agrupamento.

```python
def amostragem_agrupamento(dataset, numero_grupos, random_state = False):
    """
    Função que recebe um dataset, um valor inteiro do número de grupos desejado e um valor booleano
    para a ativação da semente de randomização. Retorna um dataframe com a amostra agrupada.
    """
    # Tamanho grupo
    tam_grupo = len(dataset) // numero_grupos
    # lista 
    grupos = []
    # Teremos 10 grupos, de IDs de 0 a 9.
    id_grupo = 0
    # Contador de grupos
    contagem = 0

    for _ in dataset.iterrows():
        if contagem > tam_grupo:
            id_grupo += 1
            contagem = 0
        grupos.append(id_grupo)
        contagem += 1
    # Nova coluna para os grupos
    dataset['grupo'] = grupos

    # Semente de randomização para manter o mesmo valor aleatório a cada rodada
    if random_state == True:        
        random.seed(7)
    # Seleciona o grupo aleatoriamente
    grupo = random.randint(0, numero_grupos)
    # Filtra um subset pelo grupo selecionado
    df_agrupamento = dataset[dataset['grupo'] == grupo]

    return df_agrupamento
        
```

```python
df_agrupamento = amostragem_agrupamento(dataset, 100, random_state = False)
df_agrupamento.shape, df_agrupamento['grupo'].value_counts()
```

## Amostragem estratificada proporcional

É uma estratégia de divisão e conquista, pois a população é dividida em grupos chamados _estratos_. Os estratos são escolhidos basedos em características da população, como sexo, edução, categoria, etc.

Após agrupados, os indivíduos são selecionados aleatoriamente para formar uma amostra. É importante que a amostra preserve a proporção de indivíduos por estrato da população.

Usamos esse tipo de amostragem quando queremos representação de todos os subgrupos da população. Contudo, necessita de um conhecimento prévio das características da população. 

```python
# Importação da biblioteca para estratificação
from sklearn.model_selection import StratifiedShuffleSplit
```

Vamos explorar o estrato rerefente à renda anual dos indivíduos, o atributo `income`.

```python
def sumario_categoria(dataset, coluna):
    """
    Função que recebe um dataset e uma coluna e retorna um dataframe com o sumário dessa coluna.
    """
    # Valor absoluto por nível da categoria
    c = dataset[coluna].value_counts(dropna=False)
    # Valor relativo por nível da categoria
    p = dataset[coluna].value_counts(dropna=False, normalize=True).mul(100).round(2)
    # concatena
    sumario = pd.concat([c,p], axis=1, keys=['counts', '%'])

    return sumario
```

```python
# Sumário da estrato income para a população
sumario_categoria(dataset, 'income')
```

Temos então que 24.08% dos indivíduos ganham mais que 50 mil dólares por ano. O complemento, 75.92%, ganha menos ou igual a 50 mil dólares por ano.

Vamos agora amostrar de forma estratificada 10% do nosso dataset.

```python
# Criação do objeto de estratificação com 10% da base de dados
split = StratifiedShuffleSplit(test_size=0.1, random_state=7)

for x, y in split.split(dataset, dataset.income):
    # População - amostra estratificada
    df_pop_menos_amostra = dataset.iloc[x]
    # Amostra estratificada
    df_amostra = dataset.iloc[y]
```

Agora veja que o tamanho da amostra corresponde a aproximadamente 10% da população. 

```python
# Quantidade de linhas do dataset VS quantidade de linhas da amostra
print(f'QTD pupulação: {dataset.shape[0]}\nQTD amostra: {df_amostra.shape[0]}')
```

```python
df_amostra.head()
```

###  Função amostragem estratificada proporcional

Para finalizar, podemos criar uma função que realize a amostragem estratificada proporcional, dado um dataset como população, o estrato referente à característica de amostragem e o tamanho da amostra.

```python
def amostragem_estratificada(dataset, estrato, tamanho_amostra):
    """
    Função que recebe um dataset, um estrato e um tamanho de amostra. 
    Retorna um dataframe com a amostra estratificada.
    """

    tam_populacao = len(dataset)
    percent_amostra = tamanho_amostra / tam_populacao

    # Criação do objeto de estratificação
    split = StratifiedShuffleSplit(test_size=percent_amostra, random_state=7)

    for x, y in split.split(dataset, dataset[estrato]):
        # População - amostra estratificada
        df_pop_menos_amostra = dataset.iloc[x]
        # Amostra estratificada
        df_amostra = dataset.iloc[y]    

    return df_amostra
```

Vamos amostrar 1500 indivíduos de acordo com o estrato de status de relacionamento, `relationship`. Primeiro, vamos conhecer a quantidade de indivíduos por nível de relacionamento. 

```python
# Sumário da estrato relationship para a amostra
sumario_categoria(dataset, 'relationship')
```

Fazendo a amostragem, temos:

```python
df_amostra_estratificada = amostragem_estratificada(dataset, 'relationship', 1500)
```

```python
# Tamanho da amostra
print(f'QTD amostra: {df_amostra_estratificada.shape[0]}')
# Sumário da estrato relationship para a amostra
sumario_categoria(df_amostra_estratificada, 'relationship')
```

##  Amostragem de reservatório

```python
def amostragem_reservatorio(dataset, amostra):
    stream = []
    for i in range(len(dataset)):
        stream.append(i)
    i = 0
    tamanho = len(dataset)

    reservatorio = [0] * amostra
    for i in range(amostra):
        reservatorio[i] = stream[i]
    
    while i < tamanho:
        j = random.randrange(i + 1)
        if j < amostra:
            reservatorio[j] = stream[i]
        i += 1

    return dataset.iloc[reservatorio]
```
